##  Total count of Brand's in women's category

```python
women_brandscount = products[products["Category"] == "Women"]["Brand Name"].value_counts().reset_index()
women_brandscount.columns = ["Brands Name", "Total Count"]
women_brandscount
```

## Count of Products in women's category

```python
women_productscount = products[products["Category"] == "Women"]["Product Name"].value_counts().reset_index()
women_productscount.columns = ["Product Name","Total Count"]
women_productscount
```

## Count of Ratings of different brands in women's category

```python
women_products = products[products["Category"] == "Women"]
women_productsrating  = women_products.pivot_table(index = "Brand Name",columns = "Ratings",aggfunc="size",fill_value=0)
women_productsrating = women_productsrating.sort_values(by=5,ascending = False)
women_productsrating["Total Ratings"] = women_productsrating.sum(axis=1)
women_productsrating = women_productsrating.sort_values("Total Ratings",ascending = False)
women_productsrating
```

## Analysis of Order's Data

```python
orders
```

```python
orders["Month"] = orders["Date"].dt.strftime("%B")
orders
```

```python
orders["Year"] = orders["Date"].dt.strftime("%Y")
orders
```

```python
orders["Total Price"] = orders["Original Price"] - (orders["Original Price"] * orders["Discount%"])
orders
```

```python
year_grp = orders.groupby("Year").agg({"Total Price":"sum"})
year_grp = year_grp.sort_values(by="Total Price",ascending=False)
year_grp
```

```python
sns.barplot(x=year_grp.index,y=year_grp["Total Price"])
plt.show()
```

```python
group = orders.groupby("Month").agg({"Total Price":"sum"})
group = group.sort_values(by="Total Price",ascending=False)
group
```

```python
plt.figure(figsize=(12,4))
sns.barplot(x=group.index,y=group["Total Price"])
plt.show()
```

## Merge of  Orders table and Products table

```python
df = pd.merge(left = orders, right = products, on = "Product ID", how = "inner")
df
```

## Average Price per Brand in all Category

```python
avg_price_brand = df.groupby(["Category","Brand Name"])["Total Price"].mean().reset_index()
avg_price_brand.columns = ["Category","Brand Name","Avg Price"]
avg_price_brand
```

```python
grp = df.groupby("Category").agg({"Total Price":"sum"})
grp = grp.sort_values(by="Total Price",ascending = False)
grp
```

```python

plt.pie(grp["Total Price"], labels=grp.index,autopct="%1.2f%%")
plt.show()
```

## Average discount per category & brand

```python
avg_discount_category_brand = df.groupby(["Category","Brand Name"])["Discount%"].mean().mul(100).round(2)
avg_discount_category_brand
```

## Average discount per product

```python
avg_discount_product = df.groupby(["Category","Product Name"])["Discount%"].mean().mul(100).round(2)
avg_discount_product
```

## Effect of discounts on sales volume (Correlation Analysis)

```python
discount_corr = df["Discount%"].corr(df["Total Price"])
discount_corr

# A slight negative correlation suggests that higher discounts slightly increase sales volume, 
#but the impact is not very strong.
```

## Best Selling products by quantity and revenue

```python
product_sales = df.groupby("Product Name").agg(qty_sold=("Order ID","count"),total_rev=("Total Price","sum"))
product_sales = product_sales.sort_values(by="total_rev",ascending=False)
product_sales
```
